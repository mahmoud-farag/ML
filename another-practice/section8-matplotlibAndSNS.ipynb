```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(10 , 100 , 100)
figure , axis =  plt.subplots()
axis.plot(x , x**2);
```

### X , y must be numpy.ndarray any thing else will not polt 

#### Scatter plot

```python
figure2, axis2  = plt.subplots()
x2 = np.linspace(0,20 , 100)
axis2.scatter(x2 , x2**2);
axis2.set(title = 'Graph',
         xlabel = 'X-axis',
        ylabel = 'y-axis');
```

#### bar plot

```python
dictionary = {"Bananas": 88,
              "Orange" : 60,
              "Apple" : 99}

figure3 , axis3 = plt.subplots()

axis3.bar(dictionary.keys(), dictionary.values());
```

#### Saving the Graphe as an Image

```python
figure3.savefig('imgs/barchart.png')
```
